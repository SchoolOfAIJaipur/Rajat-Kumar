```python
import random as rand
import numpy as num
e=2.71
def GPDF(sig,x,u):
    return(1/(((2*num.pi)**0.5)*sig)*e**-((x-u)*(x-u))/(sig*sig))
randomX=[]
randomY=[]
for i in range(0,1000):
    randomX.append(rand.randint(0,1000))
for i in range(0,len(randomX)):
    randomY.append(GPDF(26,randomX[i],28))
print(randomY)
```
