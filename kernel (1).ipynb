```python
import matplotlib.pyplot as plt

def LineEqn(m,x,c):
    return(m*x+c)
randomX=[]
randomY=[]
for i in range(0,1000):
    randomX.append(rand.randint(0,1000))
    
for i in range(0,len(randomX)):
    randomY.append(LineEqn(2,randomX[i],2))
plt.plot(randomX,randomY)
plt.show()    
```

```python
import matplotlib.pyplot as plt
def PabEqn(x):
    return(x*x)
randomX=[]
randomY=[]

for i in range(-1000,1000):
    randomX.append(rand.randint(-1000,1000))
    
for i in range(0,len(randomX)):
    randomY.append(PabEqn(randomX[i]))
    
plt.plot(randomX,randomY)
plt.show()
```

```python
import matplotlib.pyplot as plt

def HypEqn(c,x):
    return((c**2)/x)
randomX=[]
randomY=[]

for i in range(0,1000):
    randomX.append(rand.randint(1,1000))
for i in range(0,len(randomX)):
    randomY.append(HypEqn(4,randomX[i]))
    
plt.plot(randomX,randomY)
plt.show()
```

```python
import matplotlib.pyplot as plt

def Elp(a,b,x):
    return(((a**2)*(b**2)-(b**2)*(x**2)/(a**2))**0.5)
randomX=[]

randomY=[]
for i in range(0,1000):
    randomX.append(rand.randint(0,1000))
for i in range(0,len(randomX)):
    randomY.append(Elp(4,3,randomX[i]))
plt.plot(randomX,randomY)
plt.show()
    
```
